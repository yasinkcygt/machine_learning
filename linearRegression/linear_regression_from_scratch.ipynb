```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training data: house sizes (in 1000 sqft)
x_train = np.array([
    1.0, 1.2, 1.5, 1.7, 2.0,
    2.2, 2.5, 2.7, 3.0, 3.2,
    3.5, 3.7, 4.0, 4.2, 4.5,
    4.7, 5.0, 5.2, 5.5, 5.7
])

# Training data: house prices (in 1000 dollars)
y_train = np.array([
    300, 340, 380, 400, 450,
    480, 520, 550, 600, 620,
    670, 700, 750, 780, 820,
    850, 900, 920, 970, 1000
])

# Print input arrays
print(f'x_train: {x_train}')
print(f'y_train: {y_train}')

print()
print()

# Print shapes (number of training examples)
print(x_train.shape)
print(y_train.shape)
```

```python
def compute_output_model(x, w, b):
    m = x.shape[0]                    # number of training examples
    f_wb = np.zeros(m)                # initialize predictions array

    for i in range(m):
        f_wb[i] = np.dot(w, x[i]) + b  # linear model: y = w*x + b

    return f_wb                        # return predicted values

w = 170
b = 100

f_wb_predict = compute_output_model(x_train, w, b)  # compute predictions
print(f_wb_predict)                                 # print predictions
```

```python
import matplotlib.pyplot as plt
```

```python
# Scatter plot: actual house prices
plt.scatter(x_train, y_train, c='b', label='actual values')

# Plot predicted prices (linear regression line)
plt.plot(x_train, f_wb_predict, c='r', label='predict line')

# Add title and axis labels
plt.title('House Pricing', c='y')
plt.xlabel('House Size', c='y')
plt.ylabel('Prices', c='y')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
def compute_cost(y_predict, y_actual): 
    """
    Computes the cost function for linear regression.
    
    Args:
      y_predict (ndarray (m,)): Predicted values from the model
      y_actual  (ndarray (m,)): True target values
    
    Returns:
      total_cost (float): Mean squared error cost: 1/(2*m) * sum((y_predict - y_actual)^2)
    """

    m = y_actual.shape[0]        # number of examples
    cost_sum = 0                 # initialize cost sum

    for i in range(m):
        cost = (y_predict[i] - y_actual[i]) ** 2  # squared error for each example
        cost_sum += cost                            

    total_cost = 1/(2 * m) * cost_sum  # compute average cost
    return total_cost
```

```python
print(compute_cost(f_wb_predict, y_train))
```
